#### Utility notebook to process Makhzan corpus, shuffle, and downsize for finetuning purposes.

```python
import pandas as pd
pd.set_option('display.max_colwidth', None) 
```

```python
# Start with the training dataset.
train_data = pd.read_json('./data/train.jsonl', lines=True)
train_data.head()
```

```python
# One cleaning step we can do is to remove very long rows in order to avoid long context issues.
# We define this in terms of the number of words in the string.
# This is in addition to dropping any empty rows.
def get_string_length(s):
    if isinstance(s, str):
        words = s.split()
        return len(words)
    return 0

train_data['text_length'] = train_data['text'].apply(get_string_length)
print(train_data['text_length'].describe())
train_data.head()
```

```python
train_data = train_data[train_data['text_length'] < 500]  # Keep rows with less than 500 words
train_data = train_data[train_data['text_length'] > 5]  # Drop empty rows or very short rows
print(len(train_data))
```

```python
#63713 examples left after cleaning, let's shuffle and save 1000 for training.
train_data = train_data.sample(frac=1).reset_index(drop=True)  # Shuffle
train_data = train_data.drop(columns=['text_length'])  # Drop the length column
train_data = train_data[:1000]  # Take the first 1000 examples
train_data.to_json('./data/training_urdu.jsonl', orient='records', lines=True, force_ascii=False)
train_data.head()
```

```python
# Repeat the above for validation and testing datasets.
# We will use a validation set of 200 examples and a test set of 100 examples.
valid_data = pd.read_json('./data/valid.jsonl', lines=True)
test_data = pd.read_json('./data/test.jsonl', lines=True)
valid_data['text_length'] = valid_data['text'].apply(get_string_length)
test_data['text_length'] = test_data['text'].apply(get_string_length)

valid_data = valid_data[valid_data['text_length'] < 500]  # Keep rows with less than 500 words
valid_data = valid_data[valid_data['text_length'] > 5]  # Drop empty rows or very short rows
print(len(valid_data))
valid_data = valid_data.sample(frac=1).reset_index(drop=True)  # Shuffle
valid_data = valid_data.drop(columns=['text_length'])  # Drop the length column
valid_data = valid_data[:200]  # Take the first 200 examples
valid_data.to_json('./data/validation_urdu.jsonl', orient='records', lines=True, force_ascii=False)
valid_data.head()

test_data = test_data[test_data['text_length'] < 500]  # Keep rows with less than 500 words
test_data = test_data[test_data['text_length'] > 5]  # Drop empty rows or very short rows
print(len(test_data))
test_data = test_data.sample(frac=1).reset_index(drop=True)  # Shuffle
test_data = test_data.drop(columns=['text_length'])  # Drop the length column
test_data = test_data[:200]  # Take the first 200 examples
test_data.to_json('./data/test_urdu.jsonl', orient='records', lines=True, force_ascii=False)
test_data.head()
```
